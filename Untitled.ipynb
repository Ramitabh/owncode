```python
#load dataset from ucsi website
import pandas as pd
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data', header=None)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
X = df.loc[:, 2:].values
y = df.loc[:, 1].values

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
le.transform(['M', 'B'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
#scikit-learn transformers
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
#estimator
from sklearn.linear_model import LogisticRegression

from sklearn.pipeline import Pipeline
```

```python
pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('pca', PCA(n_components=2)),
                    ('clf', LogisticRegression(random_state=1))])
pipe_lr.fit(X_train, y_train)
print('Test Accuracy: %.3f' % pipe_lr.score(X_test, y_test))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('clf', LogisticRegression(penalty='l2', random_state=0))])
train_sizes, train_scores, test_scores =learning_curve(estimator=pipe_lr, 
                      X=X_train, 
                      y=y_train, 
                      train_sizes=np.linspace(0.1, 1.0, 10), 
                      cv=10,
                       n_jobs=1)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8, 5) 
plt.rcParams['axes.titlesize'] = 'large'


plt.plot(train_sizes, train_mean, 
         color='blue', marker='o', 
         markersize=5, 
          label='training accuracy')
plt.fill_between(train_sizes, 
                 train_mean + train_std,
                 train_mean - train_std, 
                 alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean, 
          color='green', linestyle='--', 
          marker='s', markersize=5, 
         label='validation accuracy')
plt.fill_between(train_sizes, 
                 test_mean + test_std,
                  test_mean - test_std, 
                  alpha=0.15, color='green')
plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim([0.9, 1.0])
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression

# Assuming X_train and y_train are already defined
X_train2 = X_train[:, [4, 14]]

# Define your cross-validation strategy using StratifiedKFold
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=1)

# Define your pipeline
pipe_lr = Pipeline([
    ('scl', StandardScaler()),
    ('pca', PCA(n_components=2)),
    ('clf', LogisticRegression(penalty='l2', random_state=0, C=100.0))
])

# Now you can use this cv object with your pipeline for cross-validation
for train_index, test_index in cv.split(X_train, y_train):
    X_train_fold, X_test_fold = X_train[train_index], X_train[test_index]
    y_train_fold, y_test_fold = y_train[train_index], y_train[test_index]
    
    # Fit your pipeline on the training fold
    pipe_lr.fit(X_train_fold, y_train_fold)
    
    # Evaluate your pipeline on the test fold
    accuracy = pipe_lr.score(X_test_fold, y_test_fold)
    
    # Print or store the evaluation result as needed
    print(f"Accuracy: {accuracy}")

# Optionally, you can use cross_val_score for a simpler evaluation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(pipe_lr, X_train, y_train, cv=cv)
print("Cross-validated Accuracy:", scores.mean())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from scipy import interp

fig = plt.figure(figsize=(7, 5))
mean_fpr = np.linspace(0, 1, 100)
all_tpr = []

# Initialize the StratifiedKFold object
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=1)

# Iterate over the folds using split method
for i, (train, test) in enumerate(cv.split(X_train2, y_train)):
    probas = pipe_lr.fit(X_train2[train], y_train[train]).predict_proba(X_train2[test])
    fpr, tpr, thresholds = roc_curve(y_train[test], probas[:, 1], pos_label=1)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=1, label='ROC fold %d (area = %0.2f)' % (i + 1, roc_auc))

    # Interpolate to get mean tpr
    mean_tpr = interp(mean_fpr, fpr, tpr)
    mean_tpr[0] = 0.0
    all_tpr.append(mean_tpr)

# Plot the random guessing line
plt.plot([0, 1], [0, 1], linestyle='--', color=(0.6, 0.6, 0.6), label='random guessing')

# Compute mean ROC curve and AUC
mean_tpr = np.mean(all_tpr, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
plt.plot(mean_fpr, mean_tpr, 'k--', label='mean ROC (area = %0.2f)' % mean_auc, lw=2)

# Plot perfect performance line
plt.plot([0, 0, 1], [0, 1, 1], lw=2, linestyle=':', color='black', label='perfect performance')

# Set plot limits and labels
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operator Characteristic')
plt.legend(loc='lower right')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, auc

# Assuming X_train, X_test, y_train, y_test are your training and testing data

# Step 1: Data preprocessing
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 2: Feature Engineering (Example: PCA for dimensionality reduction)
pca = PCA(n_components=10)  # Adjust the number of components based on your data
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Step 3: Model selection and hyperparameter tuning (Example: Random Forest)
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

rf = RandomForestClassifier(random_state=0)
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_pca, y_train)

best_rf = grid_search.best_estimator_

# Step 4: Evaluate the model
y_pred = best_rf.predict(X_test_pca)

accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, best_rf.predict_proba(X_test_pca)[:, 1])

print(f"Accuracy: {accuracy:.3f}")
print(f"ROC AUC: {roc_auc:.3f}")

# Optional: Plot ROC curve
fpr, tpr, _ = roc_curve(y_test, best_rf.predict_proba(X_test_pca)[:, 1])
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

#  Accuracy of 0.947 and an ROC AUC of 0.991 are indeed very good results for a classification model. Here's a breakdown of what these scores mean and why they are considered impressive:

Accuracy (0.947):
Accuracy represents the proportion of correctly classified instances out of the total instances.
An accuracy of 0.947 means that approximately 94.7% of the test instances were classified correctly by the model.
This is a high accuracy score and indicates that the model is performing very well on the test data.
ROC AUC (0.991):
ROC AUC (Receiver Operating Characteristic Area Under Curve) is a measure of the model's ability to distinguish between the classes.
An ROC AUC of 0.991 suggests that the model has an excellent ability to differentiate between the positive and negative classes.
A score close to 1 (like 0.991) indicates strong performance, with very few false positives and false negatives.
